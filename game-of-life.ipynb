Consider a grid of 2D cells each represented by a 2D unit square. Each cell has two states: dead (encoded as 0) or alive (encoded as 1). Each cell in the grid is updated in discrete time steps depending on the **state of its eight neighbors**. 

Initially, some cells are selected randomly to be alive and the rest are dead. The grid is updated in **each time step** according to the following **basic** rules:
- if a dead cell has 3 neighbors alive, then it the comes back to life
- if an alive cell has 2 or 3 neighbors alive, then the cell continues to be alive.
- if an alive cell has less than 2 or more than 3 neighbors alive, then the cell dies.

**The goal** of the game is to find patterns that evolve in interesting ways â€“ something that people have now been doing for over 50 years.

````python
import numpy as np
import matplotlib.pyplot as plt

def init_life_state_1(n, m, p):
    """
    Generate an initial random subset of life cells (2D points)
    IN: n, int, number of rows
        m, int, number of columns
        p, float, probability of a cell being alive



    OUT: ndarray of shape (n, m), initial state of the cells
    """
    return np.random.choice([0, 1], size=(n, m), p=[1 - p, p])

def draw_life_state_1(life_state):
    """
    Display the 2D positions of the selected collection of cells (2D points)
    IN: life_state, ndarray of shape (n, m), initial state of the cells
    OUT: None
    """
    plt.figure(figsize=(10, 10))
    n, m = life_state.shape
    plt.xlim(-0.5, m - 0.5)
    plt.ylim(-0.5, n - 0.5)
    plt.gca().set_aspect('equal', adjustable='box')

    alive_cells = np.argwhere(life_state == 1)
    for cell in alive_cells:
        x, y = cell[1], cell[0]
        plt.fill([x-0.5, x-0.5, x+0.5, x+0.5], [y-0.5, y+0.5, y+0.5, y-0.5], color='black')

    plt.gca().invert_yaxis()
    plt.axis('off')
    plt.show()
````

##### Task 1.2

Implement a Python function `update_life_state_1()` that takes a 2D NumPy array representing the current state of the grid and returns a new 2D NumPy array representing the next state of the grid according to the rules specified above.

Note: Make sure that all the cells are updated simultaneously. That is, the new state of each cell should be computed based on the original state of all the cells.

````python
import numpy as np

def update_life_state_1(life_state, out_life_state=None):
    """
    For each cell, evaluate the update rules independently to obtain its new state.
    IN: life_state, ndarray of shape (n, m), initial state of the cells
        out_life_state, ndarray of shape (n, m), for storing the next state of the cells. If None, create a new array.
    OUT: ndarray of shape (n, m), next state of the cells
    """
    n, m = life_state.shape

    # If no output array is provided, create a new one
    if out_life_state is None:
        out_life_state = np.zeros((n, m), dtype=int)

    # Iterate through each cell in the grid
    for i in range(n):
        for j in range(m):
            # Calculate the number of live neighbors
            alive_neighbors = 0
            for x in range(-1, 2):  # Iterate over neighbor rows (-1, 0, 1)
                for y in range(-1, 2):  # Iterate over neighbor columns (-1, 0, 1)
                    if x == 0 and y == 0:
                        continue  # Skip the current cell itself

                    ni, nj = i + x, j + y
                    # Check if the neighbor is within bounds
                    if 0 <= ni < n and 0 <= nj < m:
                        alive_neighbors += life_state[ni, nj]

            # Apply Game of Life rules based on the count of alive neighbors
            if life_state[i, j] == 1:
                # Rule 2: Any live cell with 2 or 3 live neighbors stays alive
                if alive_neighbors == 2 or alive_neighbors == 3:
                    out_life_state[i, j] = 1
                else:
                    out_life_state[i, j] = 0  # Dies due to underpopulation or overpopulation
            else:
                # Rule 1: Any dead cell with exactly 3 live neighbors becomes alive
                if alive_neighbors == 3:
                    out_life_state[i, j] = 1
                else:
                    out_life_state[i, j] = 0  # Stays dead

    return out_life_state
````

##### Task 1.4

Implement a Python function `play_game_of_life_1()` that uses the functions implemented above to play the game of life. The function should ask the user the desired number of iterations to update the grid and display the grid at each iteration. When reaching the desired number of iterations, the function should ask the user if he/she wants to update more times. If the user does not want to continue updating, the function should ask the user if he/she wants to save the initial collection of life cells and the final configuration (caveat: this may not be always feasible) as a csv file.

````python
import csv

def play_game_of_life_1():
    """
    Play the Game of Life interactively.
    IN: None
    OUT: None
    """
    # Step 1: Ask user for initial state setup method
    init_choice = input("Do you want to (1) generate a random state or (2) load from a CSV file? (Enter 1 or 2): ").strip()

    if init_choice == '1':
        n = int(input("Enter the number of rows: "))
        m = int(input("Enter the number of columns: "))
        p = float(input("Enter the probability of a cell being alive (0 to 1): "))
        initial_state = init_life_state_1(n, m, p)
    elif init_choice == '2':
        file_name = input("Enter the CSV file name to load the initial state (e.g., initial_state.csv): ")
        initial_state = load_life_state_from_csv(file_name)
        n, m = initial_state.shape
    else:
        print("Invalid input. Exiting...")
        return

    current_state = initial_state.copy()
    print("\nInitial State:")
    draw_life_state_1(current_state)

    # Step 2: Ask user for number of iterations
    iterations = int(input("\nEnter the number of iterations: "))

    # Step 3: Simulate the game
    for i in range(iterations):
        current_state = update_life_state_1(current_state)
        print(f"\nState after {i + 1} iteration(s):")
        draw_life_state_1(current_state)

    # Step 4: Ask if user wants to update more times
    while True:
        more_updates = input("Do you want to update more times? (yes/no): ").strip().lower()
        if more_updates == 'yes':
            more_iterations = int(input("Enter the number of additional iterations: "))
            for i in range(more_iterations):
                current_state = update_life_state_1(current_state)
                print(f"\nState after {i + 1} additional iteration(s):")
                draw_life_state_1(current_state)
        elif more_updates == 'no':
            break
        else:
            print("Invalid input. Please enter 'yes' or 'no'.")

    # Step 5: Ask if user wants to save the initial and final states as CSV files
    save_data = input("Do you want to save the initial and final configurations as CSV files? (yes/no): ").strip().lower()
    if save_data == 'yes':
        initial_file_name = input("Enter the file name to save the initial state (e.g., initial_state.csv): ")
        final_file_name = input("Enter the file name to save the final state (e.g., final_state.csv): ")
        with open(initial_file_name, mode='w', newline='') as file:
            writer = csv.writer(file)
            for row in initial_state:
                writer.writerow(row)
        with open(final_file_name, mode='w', newline='') as file:
            writer = csv.writer(file)
            for row in current_state:
                writer.writerow(row)
        print(f"Initial state saved as '{initial_file_name}' and final state saved as '{final_file_name}'.")
    else:
        print("Data not saved. Exiting...")

# Run the game
play_game_of_life_1()

def load_life_state_from_csv(file_name):
    """
    Load an initial life state from a CSV file.
    IN: file_name, str, path to the CSV file
    OUT: ndarray of shape (n, m), initial state of the cells
    """
    with open(file_name, mode='r') as file:
        reader = csv.reader(file)
        life_state = np.array([list(map(int, row)) for row in reader])
    return life_state
````

##### Task 2.2

Implement a Python function `update_life_state_2()` that takes the same in input as `update_life_state_1()` plus the four integers that defaults to `b1=3, b2=3, d1=2, d2=3` and returns a new 2D NumPy array representing the next state of the grid according to the rules specified above.

````python
def update_life_state_2(life_state, b1=3, b2=3, d1=2, d2=3, out_life_state=None):
    """
    For each cell, evaluate the update rules specified above to obtain its new state.
    IN: 
        life_state (ndarray): Array of shape (n, m), initial state of the cells.
        b1 (int): Lower bound of the number of neighbors for a dead cell to come back to life.
        b2 (int): Upper bound of the number of neighbors for a dead cell to come back to life.
        d1 (int): Lower bound of the number of neighbors for an alive cell to continue to be alive.
        d2 (int): Upper bound of the number of neighbors for an alive cell to continue to be alive.
        out_life_state (ndarray): Array of shape (n, m) to store the next state of the cells. If None, a new array is created.
        
    OUT: 
        ndarray: Array of shape (n, m), representing the next state of the cells.
    """
    n, m = life_state.shape

    # If no output array is provided, create a new one
    if out_life_state is None:
        out_life_state = np.zeros((n, m), dtype=int)

    # Iterate through each cell in the grid
    for i in range(n):
        for j in range(m):
            # Calculate the number of live neighbors
            alive_neighbors = 0
            for x in range(-1, 2):  # Iterate over neighbor rows (-1, 0, 1)
                for y in range(-1, 2):  # Iterate over neighbor columns (-1, 0, 1)
                    if x == 0 and y == 0:
                        continue  # Skip the current cell itself

                    ni, nj = i + x, j + y
                    # Check if the neighbor is within bounds
                    if 0 <= ni < n and 0 <= nj < m:
                        alive_neighbors += life_state[ni, nj]

            # Apply the user-defined rules
            if life_state[i, j] == 1:
                # Alive cell stays alive if it has between d1 and d2 live neighbors (inclusive)
                if d1 <= alive_neighbors <= d2:
                    out_life_state[i, j] = 1
                else:
                    out_life_state[i, j] = 0  # Dies due to underpopulation or overpopulation
            else:
                # Dead cell becomes alive if it has between b1 and b2 live neighbors (inclusive)
                if b1 <= alive_neighbors <= b2:
                    out_life_state[i, j] = 1
                else:
                    out_life_state[i, j] = 0  # Stays dead

    return out_life_state
````
