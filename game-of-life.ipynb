Consider a grid of 2D cells each represented by a 2D unit square. Each cell has two states: dead (encoded as 0) or alive (encoded as 1). Each cell in the grid is updated in discrete time steps depending on the **state of its eight neighbors**. 

Initially, some cells are selected randomly to be alive and the rest are dead. The grid is updated in **each time step** according to the following **basic** rules:
- if a dead cell has 3 neighbors alive, then it the comes back to life
- if an alive cell has 2 or 3 neighbors alive, then the cell continues to be alive.
- if an alive cell has less than 2 or more than 3 neighbors alive, then the cell dies.

````python
import numpy as np
import matplotlib.pyplot as plt

def init_life_state_1(n, m, p):
    """
    Generate an initial random subset of life cells (2D points)
    IN: n, int, number of rows
        m, int, number of columns
        p, float, probability of a cell being alive
    OUT: ndarray of shape (n, m), initial state of the cells
    """
    return np.random.choice([0, 1], size=(n, m), p=[1 - p, p])

def draw_life_state_1(life_state):
    """
    Display the 2D positions of the selected collection of cells (2D points)
    IN: life_state, ndarray of shape (n, m), initial state of the cells
    OUT: None
    """
    plt.figure(figsize=(10, 10))
    n, m = life_state.shape
    plt.xlim(-0.5, m - 0.5)
    plt.ylim(-0.5, n - 0.5)
    plt.gca().set_aspect('equal', adjustable='box')

    alive_cells = np.argwhere(life_state == 1)
    for cell in alive_cells:
        x, y = cell[1], cell[0]
        plt.fill([x-0.5, x-0.5, x+0.5, x+0.5], [y-0.5, y+0.5, y+0.5, y-0.5], color='lightgray')

    plt.gca().invert_yaxis()
    plt.axis('off')
    plt.show()
````

##### Task 1.2

Implement a Python function `update_life_state_1()` that takes a 2D NumPy array representing the current state of the grid and returns a new 2D NumPy array representing the next state of the grid according to the rules specified above.

Note: Make sure that all the cells are updated simultaneously. That is, the new state of each cell should be computed based on the original state of all the cells.

````python
import numpy as np

def update_life_state_1(life_state, out_life_state=None):
    """
    For each cell, evaluate the update rules independently to obtain its new state.
    IN: life_state, ndarray of shape (n, m), initial state of the cells
        out_life_state, ndarray of shape (n, m), for storing the next state of the cells. If None, create a new array.
    OUT: ndarray of shape (n, m), next state of the cells
    """
    n, m = life_state.shape
    if out_life_state is None:
        out_life_state = np.zeros((n, m), dtype=int)

    # Iterate through cells
    for i in range(n):
        for j in range(m):
            alive_neighbors = 0
            for x in range(-1, 2):
                for y in range(-1, 2):
                    if x == 0 and y == 0:
                        continue
                    ni, nj = i + x, j + y
                    if 0 <= ni < n and 0 <= nj < m:
                        alive_neighbors += life_state[ni, nj]

            # Apply rules based on the count of alive neighbors
            if life_state[i, j] == 1:
                # Rule 2: Any live cell with 2 or 3 live neighbors stays alive
                if alive_neighbors == 2 or alive_neighbors == 3:
                    out_life_state[i, j] = 1
                else:
                    out_life_state[i, j] = 0
            else:
                # Rule 1: Any dead cell with exactly 3 live neighbors becomes alive
                if alive_neighbors == 3:
                    out_life_state[i, j] = 1
                else:
                    out_life_state[i, j] = 0

    return out_life_state
````

````python
def update_life_state_2(life_state, b1=3, b2=3, d1=2, d2=3, out_life_state=None):
    """
    For each cell, evaluate the update rules specified above to obtain its new state.
    IN: 
        life_state (ndarray): Array of shape (n, m), initial state of the cells.
        b1 (int): Lower bound of the number of neighbors for a dead cell to come back to life.
        b2 (int): Upper bound of the number of neighbors for a dead cell to come back to life.
        d1 (int): Lower bound of the number of neighbors for an alive cell to continue to be alive.
        d2 (int): Upper bound of the number of neighbors for an alive cell to continue to be alive.
        out_life_state (ndarray): Array of shape (n, m) to store the next state of the cells. If None, a new array is created.
        
    OUT: 
        ndarray: Array of shape (n, m), representing the next state of the cells.
    """
    n, m = life_state.shape
    if out_life_state is None:
        out_life_state = np.zeros((n, m), dtype=int)

    # Iterate through cells
    for i in range(n):
        for j in range(m):
            alive_neighbors = 0
            for x in range(-1, 2):
                for y in range(-1, 2):
                    if x == 0 and y == 0:
                        continue
                    ni, nj = i + x, j + y
                    if 0 <= ni < n and 0 <= nj < m:
                        alive_neighbors += life_state[ni, nj]
            # Apply user rules
            if life_state[i, j] == 1:
                # Alive cell stays alive if it has between d1 and d2 live neighbors
                if d1 <= alive_neighbors <= d2:
                    out_life_state[i, j] = 1
                else:
                    out_life_state[i, j] = 0
            else:
                # Dead cell becomes alive if it has between b1 and b2 live neighbors
                if b1 <= alive_neighbors <= b2:
                    out_life_state[i, j] = 1
                else:
                    out_life_state[i, j] = 0 

    return out_life_state
````

##### Task 3.2

Implement a Python function `update_life_state_3()` that takes a 2D NumPy array representing the current state of the grid, and a dictionary specifying the rules, and returns a new 2D NumPy array representing the next state of the grid according to the rules specified above.

For the Susceptible-Infected-Removed rules, the dictionary should be as follows:
```python
rules = {
    0: [  # removed cells
        {
            "turn_to": 0  # always stay removed
        }
    ],
    1: [  # susceptible cells
        {
            "neighbor_to": {
                "if": [  # adjacent to at least 1 infected cell (type 2)
                    {
                        "at_least": 1,
                        "at_most": 9,  # always true
                        "type": 2
                    }
                ],
                "then": {
                    "probability": [  # 25% chance of getting infected and 75% chance of staying susceptible
                        {
                            "value": 0.25,  # transmission rate 25%
                            "then": {
                                "turn_to": 2
                            },
                        },
                        {
                            "value": 0.75,  # leftover 75%
                            "then": {
                                "turn_to": 1
                            }
                        }
                    ]
                }
            }
        }
    ],
    2: [
        {  # infected cells
            "probability": [  # 50% chance of recovering and 50% chance of staying infected
                {
                    "value": 0.5,  # recovery rate 50%
                    "then": {
                        "turn_to": 0
                    }
                },
                {
                    "value": 0.5,  # leftover 50%
                    "then": {
                        "turn_to": 2
                    }
                }
            ]
        }
    ]
}
```

````python
import numpy as np

def update_life_state_3(life_state, rules_dict, out_life_state=None):
    """
    For each cell evaluate the update rules specified above to obtain its new state.
    IN: life_state, ndarray of shape (n, m), initial state of the cells
        rules_dict, dict, rules for updating the cells
        out_life_state, ndarray of shape (n, m), for storing the next state of the cells, if None, create a new array
    OUT: ndarray of shape (n, m), next state of the cells
    """
    n, m = life_state.shape
    if out_life_state is None:
        out_life_state = np.zeros((n, m), dtype=int)

    for i in range(n):
        for j in range(m):
            current_state = life_state[i, j]
            next_state = current_state
            state_rules = rules_dict.get(str(current_state), [])
            # Go through rules for each cell
            for rule in state_rules:
                if "neighbor_to" in rule:
                    condition = rule["neighbor_to"]["if"][0]
                    at_least = condition["at_least"]
                    at_most = condition["at_most"]
                    neighbor_type = condition["type"]

                    # Count neighbors of the specified type using a nested loop
                    neighbor_count = 0
                    for x in range(-1, 2):
                        for y in range(-1, 2):
                            if x == 0 and y == 0:
                                continue  # Skip the current cell
                            ni, nj = i + x, j + y
                            if 0 <= ni < n and 0 <= nj < m and life_state[ni, nj] == neighbor_type:
                                neighbor_count += 1

                    # Check condition and apply probability-based update
                    if at_least <= neighbor_count <= at_most:
                        probabilities = rule["neighbor_to"]["then"]["probability"]
                        rand = np.random.rand()
                        cumulative_prob = 0
                        for prob in probabilities:
                            cumulative_prob += prob["value"]
                            if rand < cumulative_prob:
                                next_state = prob["then"]["turn_to"]
                                break

                # Check for direct probability-based updates
                elif "probability" in rule:
                    probabilities = rule["probability"]
                    rand = np.random.rand()
                    cumulative_prob = 0
                    for prob in probabilities:
                        cumulative_prob += prob["value"]
                        if rand < cumulative_prob:
                            next_state = prob["then"]["turn_to"]
                            break

            # Update the cell
            out_life_state[i, j] = next_state

    return out_life_state
````

##### Task 3.4

Implement a Python function `play_game_of_life_3()` that is similar to Task 1.4 `play_game_of_life_1()`, except that the user should be able to specify the rules as a JSON file in the format of a dictionary. When saving the initial collection of life cells and the final configuration, the function should also write the rules as a JSON file in the format of a dictionary.

````python
import numpy as np
import json
import csv


def play_game_of_life_3():
    """
    Play the Game of Life 3 with custom rules specified via a JSON file.
    IN: None
    OUT: None
    """
    # Ask user for initial state setup method
    # Note that for extra credit, we implemented loading state from an existing CSV file and JSON
    init_choice = input("Do you want to (1) generate a random state or (2) load from a CSV file? (Enter 1 or 2): ").strip()

    # Random setup
    if init_choice == '1':
        n = int(input("Enter the number of rows: "))
        m = int(input("Enter the number of columns: "))
        print("Enter the probabilities for each state [Removed, Susceptible, Infected]:")
        p_list = list(map(float, input("Enter three probabilities separated by spaces (e.g., 0.2 0.6 0.2): ").split()))
        states = [0, 1, 2]
        initial_state = init_life_state_3(n, m, p_list, states)

    # Load CSV
    elif init_choice == '2':
        file_name = input("Enter the CSV file name to load the initial state (e.g., initial_state.csv): ")
        initial_state = load_state_from_csv(file_name)
        n, m = initial_state.shape
    else:
        print("Invalid input. Please restart the program and choose 1 or 2.")
        return

    # Show initial state
    current_state = initial_state.copy()
    print("\nInitial State:")
    draw_life_state_3(current_state, ['black', 'green', 'red'])

    # Load custom rules from a JSON file
    rules_file_path = input("Enter the path to the JSON file containing the rules: ")
    with open(rules_file_path, 'r') as f:
        rules = json.load(f)

    # Prompt user for the number of iterations
    iterations = int(input("Enter the number of iterations: "))
    for iteration in range(1, iterations + 1):
        print(f"\nIteration {iteration}:")
        current_state = update_life_state_3(current_state, rules)
        draw_life_state_3(current_state, ['black', 'green', 'red'])

    # Ask user for extra iterations
    while True:
        more_updates = input("Do you want to update more times? (yes/no): ").strip().lower()
        if more_updates == 'yes':
            more_iterations = int(input("Enter the number of additional iterations: "))
            for i in range(more_iterations):
                current_state = update_life_state_3(current_state, rules)
                print(f"\nState after {i + 1} additional iteration(s):")
                draw_life_state_3(current_state, ['black', 'green', 'red'])
        elif more_updates == 'no':
            break
        else:
            print("Invalid input. Please enter 'yes' or 'no'.")

    # Ask if the user wants to save the results
    save_data = input("Do you want to save the initial state, final state, and rules? (yes/no): ").strip().lower()
    if save_data == 'yes':
        # Get file names from user
        initial_csv = input("Enter the name for the CSV file to save the initial state (e.g., initial_state.csv): ")
        final_csv = input("Enter the name for the CSV file to save the final state (e.g., final_state.csv): ")
        rules_json = input("Enter the name for the JSON file to save the rules (e.g., rules.json): ")
        save_state_to_csv(initial_state, initial_csv)
        save_state_to_csv(current_state, final_csv)
        with open(rules_json, 'w') as f:
            json.dump(rules, f, indent=4)

        print(f"Initial state saved to '{initial_csv}', final state saved to '{final_csv}', and rules saved to '{rules_json}'.")
    else:
        print("Simulation results were not saved.")

# Helper function to save to CSV
def save_state_to_csv(state, file_name):
    with open(file_name, 'w', newline='') as file:
        writer = csv.writer(file)
        writer.writerows(state)
    print(f"State saved to '{file_name}'.")

# Helper function to load from CSV
def load_state_from_csv(file_name):
    with open(file_name, 'r') as file:
        reader = csv.reader(file)
        return np.array([list(map(int, row)) for row in reader])

# Run the game
# play_game_of_life_3()
````
